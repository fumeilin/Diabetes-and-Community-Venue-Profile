### CDC diabetes prevalence file: 2004-2009
https://www.cdc.gov/diabetes/atlas/obesityrisk/county_excelstatelistDM.html

```python
import pandas as pd 
import numpy as np 

#cdc file contains # of cases, age-adjusted prevalence rate of DM, age-adjusted prevalence rate of Obesity
df = pd.read_csv("cdcdownloadfile.csv")
print(df.shape)
df.head()
```

```python
df['county_code'] = df['fips_Code'].astype(str)
df['county_code'] = df['county_code'].str.zfill(5)
df.drop('county_code',axis=1,inplace=True)
df['countyid'] = '0500000US' + df['county_code']
```

```python
df['dm_aap_2009'] = df['dm_aap_2009'].astype(float)
df['ob_aap_2009'] = df['ob_aap_2009'].astype(float)
df['dm_2009'] = df['dm_2009'].astype(float)
```

```python
df.describe()
```

```python
c = pd.read_csv("venuefile.csv")
c.head()
```

```python
ds = c.merge(df[['countyid','dm_2009','dm_aap_2009','ob_aap_2009']], how='inner',on='countyid')
```

```python
ds.head()
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy import stats
```

```python
# compute age-adjusted population
d = ds[['countyid','dm_2009','dm_aap_2009','ob_aap_2009','category_c','venue_c']]
d['aa_pop_2009'] = 100 * d['dm_2009'] / d['dm_aap_2009']
```

```python
# exclude outliers and recheck scatterplots
d = d[(d['aa_pop_2009'] < 1000000) & (d['category_c'] > 20) & (d['venue_c'] > 60)]
d.describe()
```
